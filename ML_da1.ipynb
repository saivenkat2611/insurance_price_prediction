```python
class HybridInsurancePredictor(BaseEstimator):
    def __init__(self, weights=None):
        # Initialize base models with the best parameters found earlier
        self.rf_model = RandomForestRegressor(
            n_estimators=1200,
            max_depth=50,
            min_samples_leaf=12,
            min_samples_split=7
        )
        
        self.svr_model = SVR(
            C=10,
            gamma=0.1,
            kernel='rbf',
            tol=0.0001
        )
        
        self.ridge_model = Ridge(
            alpha=20,
            random_state=42
        )
        
        self.scaler = StandardScaler()
        self.weights = weights if weights is not None else [0.5, 0.3, 0.2]  # RF, SVR, Ridge
        
    def fit(self, X, y):
        # Scale the features
        X_scaled = self.scaler.fit_transform(X)
        
        # Fit individual models
        self.rf_model.fit(X_scaled, y.ravel())
        self.svr_model.fit(X_scaled, y.ravel())
        self.ridge_model.fit(X_scaled, y.ravel())
        
        return self
    
    def predict(self, X):
        # Scale the input
        X_scaled = self.scaler.transform(X)
        
        # Get predictions from each model
        rf_pred = self.rf_model.predict(X_scaled)
        svr_pred = self.svr_model.predict(X_scaled)
        ridge_pred = self.ridge_model.predict(X_scaled)
        
        # Combine predictions using weights
        final_pred = (
            self.weights[0] * rf_pred +
            self.weights[1] * svr_pred +
            self.weights[2] * ridge_pred
        )
        
        return final_pred
    
    def optimize_weights(self, X_val, y_val):
        """Optimize the weights using a validation set"""
        X_scaled = self.scaler.transform(X_val)
        
        # Get predictions from each model
        rf_pred = self.rf_model.predict(X_scaled)
        svr_pred = self.svr_model.predict(X_scaled)
        ridge_pred = self.ridge_model.predict(X_scaled)
        
        # Simple grid search for weights
        best_score = float('-inf')
        best_weights = self.weights
        
        for w1 in np.arange(0.3, 0.7, 0.1):
            for w2 in np.arange(0.2, 0.5, 0.1):
                w3 = 1 - w1 - w2
                if w3 > 0:
                    weights = [w1, w2, w3]
                    pred = (
                        weights[0] * rf_pred +
                        weights[1] * svr_pred +
                        weights[2] * ridge_pred
                    )
                    score = r2_score(y_val, pred)
                    
                    if score > best_score:
                        best_score = score
                        best_weights = weights
        
        self.weights = best_weights
        return best_weights

# Now let's test the hybrid model
# Split data into train, validation, and test sets
from sklearn.model_selection import train_test_split

# First split into train+val and test
X_temp, X_test, y_temp, y_test = train_test_split(X_, y_, test_size=0.2, random_state=42)
# Then split train+val into train and val
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.2, random_state=42)

# Initialize and train the hybrid model
hybrid_model = HybridInsurancePredictor()
hybrid_model.fit(X_train, y_train)

# Optimize weights using validation set
best_weights = hybrid_model.optimize_weights(X_val, y_val)
print("Optimized weights:", best_weights)

# Make predictions
y_pred_hybrid_train = hybrid_model.predict(X_train)
y_pred_hybrid_test = hybrid_model.predict(X_test)

# Calculate metrics
r2_score_hybrid_train = r2_score(y_train, y_pred_hybrid_train)
r2_score_hybrid_test = r2_score(y_test, y_pred_hybrid_test)
rmse_hybrid = np.sqrt(mean_squared_error(y_test, y_pred_hybrid_test))

print('\nHybrid Model Performance:')
print('R2 score (train) : {0:.3f}'.format(r2_score_hybrid_train))
print('R2 score (test)  : {0:.3f}'.format(r2_score_hybrid_test))
print('RMSE            : {0:.3f}'.format(rmse_hybrid))

# Add hybrid model results to the comparison DataFrame
models.append(('Hybrid Model', rmse_hybrid, r2_score_hybrid_train, r2_score_hybrid_test, np.mean([r2_score_hybrid_train, r2_score_hybrid_test])))
predict = pd.DataFrame(data=models, columns=['Model', 'RMSE', 'R2_Score(training)', 'R2_Score(test)', 'Cross-Validation'])

# Visualize updated results
plt.figure(figsize=(12,8))
predict.sort_values(by=['Cross-Validation'], ascending=False, inplace=True)
sns.barplot(x='Cross-Validation', y='Model', data=predict, palette='Reds')
plt.xlabel('Cross Validation Score')
plt.ylabel('Model')
plt.title('Model Comparison Including Hybrid Model')
plt.show()
```
